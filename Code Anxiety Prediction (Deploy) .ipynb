### Anova for Gender and Occupation

The ANOVA test was conducted on the Gender and Occupation columns to determine whether there was a significant difference in the average level of social anxiety between groups in each category, which will later be used to make decisions about whether these two features will be used or not in building machine learning.

```python
import scipy.stats as stats

# Ensure the target column is numeric (in case it's read as string)
df["Anxiety Level (1-10)"] = pd.to_numeric(df["Anxiety Level (1-10)"], errors='coerce')

# -------------------------------
# SHAPIRO-WILK NORMALITY TEST - GENDER
# -------------------------------
print("🔍 Shapiro-Wilk Normality Test per Gender group:")
gender_groups = df.groupby("Gender")["Anxiety Level (1-10)"]
normal_gender = True  # flag to determine if ANOVA is valid

# Loop through each gender group to test normality
for name, group in gender_groups:
    stat, p = stats.shapiro(group)
    print(f"- {name}: W={stat:.4f}, p={p:.4f}")
    if p < 0.05:
        normal_gender = False  # one group is not normal → switch to non-parametric

# -------------------------------
# ANOVA or KRUSKAL-WALLIS - GENDER
# -------------------------------
print("\n📊 Gender ANOVA/Kruskal-Wallis Result:")
samples = [group for _, group in gender_groups]  # convert groupby object to list of samples

if normal_gender:
    # If all groups are normal, use parametric ANOVA
    f_stat, p_val = stats.f_oneway(*samples)
    print(f"ANOVA → F = {f_stat:.4f}, p = {p_val:.4f}")
else:
    # If any group is non-normal, use Kruskal-Wallis (non-parametric)
    h_stat, p_val = stats.kruskal(*samples)
    print(f"Kruskal-Wallis → H = {h_stat:.4f}, p = {p_val:.4f}")


# -------------------------------
# SHAPIRO-WILK NORMALITY TEST - OCCUPATION
# -------------------------------
print("\n🔍 Shapiro-Wilk Normality Test per Occupation group:")
occupation_groups = df.groupby("Occupation")["Anxiety Level (1-10)"]
normal_occupation = True  # flag for occupation

# Loop through each occupation group
for name, group in occupation_groups:
    if len(group) >= 3:  # Shapiro requires at least 3 samples
        stat, p = stats.shapiro(group)
        print(f"- {name}: W={stat:.4f}, p={p:.4f}")
        if p < 0.05:
            normal_occupation = False
    else:
        # Skip very small groups that can't be tested
        print(f"- {name}: skipped (less than 3 samples)")

# -------------------------------
# ANOVA or KRUSKAL-WALLIS - OCCUPATION
# -------------------------------
print("\n📊 Occupation ANOVA/Kruskal-Wallis Result:")
samples = [group for _, group in occupation_groups if len(group) >= 3]

if normal_occupation:
    f_stat, p_val = stats.f_oneway(*samples)
    print(f"ANOVA → F = {f_stat:.4f}, p = {p_val:.4f}")
else:
    h_stat, p_val = stats.kruskal(*samples)
    print(f"Kruskal-Wallis → H = {h_stat:.4f}, p = {p_val:.4f}")

```



---

#### 📌  Key Insight

1. `Gender`

   * The **Kruskal-Wallis test** result gives a **p-value = 0.9449**, indicating **no statistically significant difference** in anxiety levels between gender groups (Female, Male, Other).
   * This means that gender does **not provide meaningful predictive power** for the `Anxiety Level` variable in this dataset.
   * 🔎 **Conclusion:** Including `Gender` would add unnecessary noise without improving model accuracy, and is best excluded.

2.  `Occupation`

   * While the **Kruskal-Wallis test** for `Occupation` shows a **significant p-value (0.0000)**, suggesting group differences in anxiety,
   * There are **more than 10 unique occupations**, many of which have **small sample sizes**, leading to:

     * **Sparse one-hot encoding**, which inflates feature space.
     * **Overfitting risk**, especially in small deployment scenarios.
     * **High heterogeneity in real-world applications**: The diversity of job types **beyond those present in the dataset** makes it difficult for the model to generalize well.
   * 🔎 **Conclusion:** Despite statistical significance, `Occupation` introduces too much granularity and real-world variability. This makes it **unsuitable for deployment** unless carefully grouped or engineered.

---
✅ **Recommendation**

* Focus on well-behaved, generalizable features such as lifestyle, physiological, and psychological indicators.
* If `Occupation` is to be considered, use **domain-informed grouping** (e.g., student, office-based, manual labor) to reduce heterogeneity and improve robustness.

## Correlations and Vif Score Numeric Column with target

VIF analysis was added to ensure there is no multicollinearity among numerical features, while correlation analysis was used to identify statistically significant relationships between input variables and anxiety level, thereby strengthening the feature selection process both statistically and interpretively.

### VIF Score

```python
import pandas as pd
from statsmodels.stats.outliers_influence import variance_inflation_factor
from statsmodels.tools.tools import add_constant


# PNumeric columns for VIF calculation
df_num = df.select_dtypes(include=['float64', 'int64'])

# VIF constant
x_vif = add_constant(df_num)

# Vif calculation
vif_df = pd.DataFrame()
vif_df["Feature"] = x_vif.columns
vif_df["VIF"] = [variance_inflation_factor(x_vif.values, i) for i in range(x_vif.shape[1])]

# Display VIF results
vif_df[vif_df["Feature"] != "const"]

```

### Correlations Matrix

```python
from scipy.stats import pearsonr


# Select only numerical columns
df_num = df.select_dtypes(include=['float64', 'int64'])

# Compute correlation matrix and p-values
cor_matrix = df_num.corr()
pval_matrix = pd.DataFrame(np.ones_like(cor_matrix), columns=cor_matrix.columns, index=cor_matrix.index)

# Fill the p-value matrix with real values using Pearson correlation
for col in cor_matrix.columns:
    for row in cor_matrix.index:
        if row != col:
            _, pval = pearsonr(df_num[row], df_num[col])
            pval_matrix.loc[row, col] = pval

# Format the correlation matrix with significance stars
annot_matrix = cor_matrix.copy().astype(str)
for r in cor_matrix.index:
    for c in cor_matrix.columns:
        if r != c:
            p = pval_matrix.loc[r, c]
            star = ""
            if p < 0.001:
                star = "***"
            elif p < 0.01:
                star = "**"
            elif p < 0.05:
                star = "*"
            annot_matrix.loc[r, c] = f"{cor_matrix.loc[r, c]:.2f}{star}"
        else:
            annot_matrix.loc[r, c] = ""  # leave diagonal empty for cleaner display

# Create a mask to hide the upper triangle of the matrix
mask = np.triu(np.ones_like(cor_matrix, dtype=bool))

# Plot the heatmap with annotations and significance stars
plt.figure(figsize=(12, 10))
sns.heatmap(cor_matrix, mask=mask, annot=annot_matrix, fmt='', cmap='coolwarm', square=True, linewidths=0.5)
plt.title("Correlation Matrix with Significance Stars")
plt.tight_layout()
plt.show()
```

📊 Reason for Binarizing the Target Variable (`Anxiety Level (1–10)`)

To simplify the classification task and enhance interpretability, we transformed the original target variable `Anxiety Level (1–10)` into a binary variable called `Anxiety Category`:
- `0` → **Low to Moderate Anxiety** (Anxiety Level 1–4)  
- `1` → **High Anxiety** (Anxiety Level 5–10)

This transformation is grounded in both clinical literature and behavioral representation, as illustrated in the following two figures:



🧠 Figure 1 – SUDS Clinical Classification





The **Subjective Units of Distress Scale (SUDS)** is a well-known psychological tool used to quantify anxiety levels. Based on this scale:
- Values around **10–40** are associated with *minimal to moderate anxiety or distress*, often not interfering with daily functioning.
- Values between **50–80** represent *moderate to high distress*, and values beyond **90** indicate *extreme panic* (not recorded in our dataset).

Therefore, by clinical formulation:
- **1–4** can be grouped as "low to moderate anxiety"
- **5–10** can be grouped as "high anxiety" based on escalation in severity and potential interference with functioning.

References:

Salkevicius, J., Damaševičius, R., Maskeliunas, R., & Laukienė, I. (2019). Anxiety level recognition for virtual reality therapy system using physiological signals. Electronics, 8(9), 1039. https://doi.org/10.3390/electronics8091039


---

😊 Figure 2 – Emotional Representation of Anxiety Scale



An emoji-based scale visually illustrates the **emotional interpretation** of anxiety levels:
- Levels **1 to 4** are represented by positive or neutral faces (green to light yellow), indicating lower emotional distress.
- Levels **5 to 10** shift into orange and red tones with sad or distressed faces, reflecting more **serious anxiety symptoms**.

References: 

Baylor College of Medicine. (n.d.). Introducing the Anxiety Intensity Scale (Teen Module 2: Identifying Anxiety and Staying Motivated). In LUNA: Learning to Understand and Navigate Anxiety. Diakses dari https://www.bcm.edu/research/faculty-labs/luna-learning-to-understand-and-navigate-anxiety/teen-modules/module-2-identifying-anxiety-and-staying-motivated/introducing-the-anxiety-intensity-scale

This representation aligns with real-life perception and serves as a meaningful support for dichotomizing the scale into two categories.

## Permutation Feature Importance

```python
from sklearn.inspection import permutation_importance
import matplotlib.pyplot as plt
import numpy as np

# Calculate Feature Importance
perm_result = permutation_importance(
    best_xgb,
    X_test_encoded,
    y_test_class,
    n_repeats=10,
    random_state=42,
    scoring='recall'
)

# Top 10
sorted_idx = perm_result.importances_mean.argsort()[::-1][:10]
features = np.array(X_test_encoded.columns)[sorted_idx]
means = perm_result.importances_mean[sorted_idx]

# Visualization
plt.figure(figsize=(8, 5))
bars = plt.barh(features, means, color='skyblue')
plt.gca().invert_yaxis()
plt.title("Permutation Feature Importance (Recall)")
plt.xlabel("Importance Score")
plt.grid(True)

# label
for bar, val in zip(bars, means):
    plt.text(
        val + 0.001,
        bar.get_y() + bar.get_height() / 2,
        f'{val:.3f}',
        va='center',
        fontsize=9,
        color='black'
    )

plt.tight_layout()
plt.show()
```

### 📌 **Key Insight: Permutation Feature Importance (Recall)**

1. **`Stress Level (1–10)`**

   * Dominates with importance score **0.277**—far above others.
   * Indicates that the model heavily relies on self-reported stress to identify high social anxiety.
   * 🔎 **Conclusion:** Retain and enhance this feature using validated psychometric scales.

2. **`Sleep Hours`**

   * Contributes positively (**0.003**) to recall.
   * Aligns with research linking poor sleep to anxiety symptoms.
   * 🔎 **Conclusion:** Maintain and consider objective data collection via wearable devices.

3. **Other features**

   * Have negligible importance (≤0.005).
   * 🔎 **Conclusion:** Make them optional—if used, require thoughtful engineering or analysis.

---

✅ **Recommendations (Statistical + Updated Psychological Approach)**

1. **Prioritize `Stress Level` and `Sleep Hours`**

   * Both show strong predictive power and clinical relevance.

2. **Measure coping strategies**

   * Problem-focused coping is associated with lower anxiety, while avoidant or emotion-focused coping increases distress (Fauziyyah et al., 2023; per Let et al., 2023)([IntechOpen][1], [journals.unisba.ac.id][2]).
   * 📌 **Implementation:** Include survey items based on validated questionnaires (e.g. problem-focused coping scales).

3. **Include perceived social support**

   * Peer support and social support inversely relate to anxiety levels (Hutchins et al., 2021; Chiu et al., 2020)([Hogrefe eContent][3]).
   * 📌 **Implementation:** Use instruments like MSPSS or peer support scales to measure perceived support.

4. **Assess negative peer experiences and bullying**

   * Cyberbullying and peer rejection predict social anxiety (Gu, 2024)([IntechOpen][1]).
   * 📌 **Implementation:** Add items regarding cyberbullying experiences and peer victimization.

5. **Track problematic social media use**

   * Strong correlation (r ≈ 0.44) between social anxiety and problematic social media usage (meta-analysis by Huang et al., 2024)([PubMed][4]).
   * 📌 **Implementation:** Include validated scales measuring compulsive or anxiety-driven social media behaviors.

# General Conclusion

## 🔍 **General Conclusion: Predictive Understanding of Social Anxiety**

By combining **statistical tests**, **machine learning interpretability techniques** (Permutation Importance, PDP, SHAP), and **evidence-based psychological literature**, we have developed a comprehensive and consistent understanding of the key predictors of social anxiety within the dataset.

✅ 1. **Primary Predictors: Stress and Sleep**

* **Stress Level (1–10)** emerged as the **most dominant and consistent predictor**, significantly influencing the model’s output across all interpretability approaches. High stress scores directly increase predicted social anxiety risk, in line with robust psychological research.
* **Sleep Hours** is the **second strongest feature**, with shorter sleep duration associated with increased anxiety. This was clearly supported by SHAP values, partial dependence trends, and real-world literature linking sleep and mental health.

 ✅ 2. **Behavioral Amplifiers: Caffeine, Alcohol, and Physical Activity**

* Features such as **Caffeine Intake**, **Alcohol Consumption**, and **Physical Activity** were not primary predictors but served as **amplifying factors** that can worsen anxiety outcomes, especially under high stress.
* SHAP interaction plots demonstrated that high caffeine intake intensified the negative impact of stress, reinforcing its importance in behavioral modeling.

 ✅ 3. **Secondary Indicator: Therapy Sessions**

* The frequency of **Therapy Sessions (per month)** showed a positive relationship with predicted anxiety risk but should be interpreted as a **proxy for severity or treatment engagement** rather than a causal feature.
* It remains useful for user profiling or severity assessment but must be **contextually understood**.

 ✅ 4. **Minor or Redundant Features**

* Features like `Sweating Level`, `Dizziness`, `Medication`, `Smoking`, `Heart Rate`, and `Breathing Rate` showed **minimal predictive value** and low SHAP scores.
* These features can be safely excluded in simplified versions of the model or bundled into a composite index for more advanced profiling.

 ✅ 5. **Features to Exclude or Simplify**

* Features such as `Gender` and `Occupation` were found to be **statistically insignificant (Anova) ** or overly heterogeneous (> 10 Categories from EDA):

  * `Gender` showed no significant difference in anxiety.
  * `Occupation` included too many rare categories, creating high-dimensional noise that hurts model generalization unless grouped manually.

2. **Eliminate or simplify non-informative or noisy features**
   Remove features like `Gender`, `Occupation`, and other low-impact variables to maintain model efficiency and avoid overfitting.

3. **Translate key variables into actionable interventions**
   Develop public mental health campaigns and screening tools targeting:

   * **Stress management** strategies.
   * **Sleep hygiene** education.
   * Awareness around **caffeine and alcohol’s** role in anxiety symptoms.

   Model interactions—especially between stress and caffeine—may unlock higher accuracy and better personalization in future predictive tools.
